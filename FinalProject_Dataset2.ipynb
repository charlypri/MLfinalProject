```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from tqdm import tqdm
```

```python
train_image = []
for i in tqdm(range(10000)):
    img = image.load_img('DATASET/TRAIN/O/O_'+str(i+1)+'.jpg', target_size=(50, 50, 3), grayscale=False)
    img = image.img_to_array(img)
    train_image.append(img)
for i in tqdm(range(10000)):
    img = image.load_img('DATASET/TRAIN/R/R_'+str(i+1)+'.jpg', target_size=(50, 50, 3), grayscale=False)
    img = image.img_to_array(img)
    train_image.append(img)
```

```python
y = [] 
print(len(train_image))
for i in range(len(train_image)):
    if i < len(train_image)/2:
        y.append(0)
    else:
        y.append(1)
```

```python
y = to_categorical(y)
X = np.array(train_image)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
X.shape
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(50, 50, 3)))#242, 208, 3 is original shape
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=3, validation_data=(X_test, y_test))
```

```python
test_image = []
for i in tqdm(range(1000)):
    img = image.load_img('DATASET/TEST/O/O_'+str(i+12568)+'.jpg', target_size=(50, 50, 3), grayscale=False)
    img = image.img_to_array(img)
    test_image.append(img)
for i in tqdm(range(1000)):
    img = image.load_img('DATASET/TEST/R/R_'+str(i+10000)+'.jpg', target_size=(50, 50, 3), grayscale=False)
    img = image.img_to_array(img)
    test_image.append(img)
X_test = np.array(test_image)
```

```python
prediction = model.predict_classes(X_test)
```

```python
id = range(len(X_test))
y_real = [] 

for i in id:
    if i < len(X_test)/2:
        y_real.append(0)
    else:
        y_real.append(1)

results = pd.DataFrame({'id':id, 'y_real': y_real, 'prediction': prediction})
```

```python
ctr =0
for i in range(len(X_test)):
    if int(results['y_real'][i]) == (results['prediction'][i]) :
        ctr = ctr +1
print("The accuracy on the Test set is ", "{0:.2f}".format(ctr/len(X_test)*100) ,"%")
```
